# Testataufgabe: Suchen und Ersetzen (25 Punkte)

Erstellen Sie ein Jupyter Notebook, welches vom Nutzer einen String einliest und in diesem den ersten nicht wiederholenden Buchstaben findet.

Stellen Sie sicher, dass der Nutzer keinen String eingeben kann, der andere Zeichen außer Kleinbuchstaben enthält und das der eingelesene String nicht verändert wird!
Verwenden Sie in dieser Aufgabe **keine anderen Operationen der Klasse string** als *length()* und *at()*!

Das Ergebnis soll genau so, wie im Beispiel ausgegeben werden, insbesondere soll hierbei ein String ausgegeben werden, in dem der erste nicht wiederholende Buchstabe in einen Großbuchstaben umgewandelt wurde.
Falls kein nicht wiederholender Buchstabe im String vorkommt, soll dies mit einer kurzen Nachricht angezeigt werden.

*Beispiel:*

    Eingabestring: rentnerin
    
    Ausgabe: 
    Originalstring: rentnerin
    Buchstabe: t
    Index: 3
    Ergebnisstring: renTnerin

***
    
## Anforderungen:
- Einlesen entsprechend der Aufgabenstellung.
- Anzahl jedes im String vorkommenden Zeichens feststellen.
- Ersten nicht wiederholenden Buchstaben identifizieren und in Großbuchstaben umwandeln.
- Ausschließliche Verwendung der Operationen *length()* und *at()* der Klasse string.
- Ausgabe wie im Beispiel.

```c++
#include <iostream>
using namespace std;

string eingabe, ausgabe;
bool erlaubt, einmalig = false;


do { 
    /* Diese Integer müssen jedes mal neu mit 0 initiert werden,
    um nicht die Werte vorheriger Werte durch unzulässige Strings anzunehmen */
    int stelle = 0;
    int alphabet[26] = {0};
    
    cout << "\nGeben Sie ein Wort oder String, ausschließlich mit kleingeschrieben lateinischen Buchstaben und ohne weitere Zeichen, ein. \n " << endl;
    cin >> eingabe;    
    /* Die For Schleife filtert unerlaubte Zeichen durch ASCI Code Nummer raus (wieso nicht Leerzeichen?)
    Das Leerzeichen hat den ASCI Code 32, lässt sich allerdings nicht ausschließen, 
    auch nicht durch eine For Schleife vor der while Schleife, da schon bei der Eingabe durch ein Leerzeichen
    die Strings getrennt werden. Die werden dann nacheinander ausgeführt, wie zwei nacheinander getätigte Eingaben, 
    sofern die erste nicht gültig ist. Wenn der erste Sting gültig ist, ist der Durchlauf komplett beendet.) */
    for (int pos = 0; pos < eingabe.length(); pos++) {    
        if (eingabe.at(pos) > 96 && eingabe.at(pos) < 123) {
            // der Array "alphabet" wird mit der Anzhal der vokommenden Buchstaben gefüllt
            int ord = eingabe.at(pos) - 'a';
            alphabet[ord]++;
        } else {
            // Damit die unten folgende While Schleife nicht beginnt, wird auch "int stelle" hochgesetzt
            pos, stelle = eingabe.length();
        }    
    }

    /* Wenn jeder Buchstabe des Wortes überprüft und kategorisiert wurde,
    wird die Stelle des Buchstabens, welcher als erstes nur einmal verwendet wird, gemerkt. */
    while (einmalig == false && stelle < eingabe.length()) {
        int bstabe = eingabe.at(stelle) - 'a';
        // <Bedingung> ? <true> : <false>
        alphabet[bstabe] == 1 ? einmalig = true : stelle++;
    }  
    if (einmalig == false) {
        cout << "\nVersuchen Sie es mit einem anderen Wort." << endl; 
        //Hier könnte auch stelle = 0; hinzugefügt werden, dann kann es globale Variable sein
    } else {      
        // Der Ergebnisstring wird erstellt
        ausgabe = eingabe;
        char alt = eingabe.at(stelle);
        char neu = eingabe.at(stelle) - 'a'+ 'A';
        for (int i = 0; i < ausgabe.length(); i++) {
            if (ausgabe[i] == alt) {
                ausgabe[i] = neu;
            }
        }
        cout << "\n\nOriginalstring: " << eingabe << "\nBuchstabe: " << eingabe.at(stelle) << "\nIndex: " << stelle << "\nErgebnisstring: " << ausgabe << endl;
        erlaubt = true;
    }
    if (erlaubt == false) {
        cout << "\nBitte befolgen Sie die Anweisung:" << endl;
    }  
} while (erlaubt == false);
```
